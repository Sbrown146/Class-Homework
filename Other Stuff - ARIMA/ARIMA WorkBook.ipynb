```python
import pandas as pd
from  pandas import datetime
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from scipy import stats
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima_model import ARIMA
import itertools
import warnings
warnings.filterwarnings('ignore')
```

```python
sales_data = ("Resources/sales-cars.csv")
sales_df = pd.read_csv(sales_data, encoding = "utf-8", parse_dates=[0], date_parser=parser)
sales_df["Months"] = np.arange(1,37,1)
sales_df1 = pd.DataFrame(sales_df) #SET AS DATAFRAME FOR INDEXING TO WORK PROPERLY
sales_df = sales_df[["Months", "Sales"]]
sales_df_months = sales_df["Months"]
sales_df_sales = sales_df["Sales"]
```

```python
#This works best with 2 defined sets for the x and y axis
plt.title("Car Sales Data")
plt.plot(sales_df)
plt.show
```

```python
#This works much better
plt.title("Car Sales Data")
plt.xlabel("Total Months")
plt.ylabel("Car Sales")
plt.legend(loc="lower right")
plt.plot(sales_df_months, sales_df_sales, marker="o", color="blue", label="Car Sales")
```

```python
#Autoregressive and Moving Average Anaylsis
#Yuck
sales_df1 = sales_df.set_index("Months")
k1, p1 = stats.normaltest(sales_df1)
print(k1)
print(p1)
print(plt.scatter(np.arange(1,37,1), sales_df1, facecolors="red", edgecolors="black"))
print(plot_acf(sales_df1))
print(plot_pacf(sales_df1))
```

```python
#Shifting
sales_df_shift1 = sales_df1.shift(1)
sales_df_shift2 = sales_df1.shift(2)
print(sales_df_shift1.head())
print(sales_df_shift2.head())
```

```python
#Differencing
sales_diff = sales_df1.diff(periods=1)
sales_diff = sales_diff[1:]
month_diff = np.arange(2,37,1)
print(sales_diff.head())
plt.title("Sales Data After First Differencing")
plt.xlabel("Months")
plt.ylabel("First Difference Sales Data")
plt.plot(sales_diff, marker="o")
```

```python
#Scatter Plot looks much better - NormalTest is Pearson test statisic p value.
#Try order 1 for differencing.
plt.title("First Difference Car Sales Data")
plt.xlabel("Months")
plt.ylabel("First Differencing for Sales Data")
plt.scatter(month_diff, sales_diff, marker="o", facecolors="red", edgecolors="black")
k2, p2 = stats.normaltest(sales_diff)
print(k2)
print(p2)
```

```python
#Second Order Differencing.  Try both out.
sales_diff2 = sales_df1.diff(periods=2)
sales_diff2 = sales_diff2[2:]
k3, p3 = stats.normaltest(sales_diff2)
print(k2)
print(p3)
print(sales_diff2.head())
plt.scatter(np.arange(3,37,1), sales_diff2, facecolors="red", edgecolors="black")
```

```python
plt.plot(sales_diff2, marker="o", color="blue", label="2nd Order Differenced Sales Data")
```

```python
#Split model for prediction purposes:
model_data = sales_df1[0:27]
predict_data = sales_df1[26:]
model_ar = AR(model_data)
model_ar_fit = model_ar.fit()
predictions = model_ar_fit.predict(start=27, end=36)
plt.plot(predict_data, color="blue")
plt.plot(predictions, color="red")
```

```python
mean_squared_error(predict_data, predictions)
```

```python
#ARIMA models 
model_arima = ARIMA(model_data, order = (9,2,0))
model_arima_fit = model_arima.fit()
predictions = model_arima_fit.forecast(steps= 9)[0]
predict_data["Months"] = np.arange(0,10,1)
predict_data = pd.DataFrame(predict_data)
predict_data = predict_data.set_index("Months")
print(predict_data)
print(predictions)
```

```python
plt.plot(predict_data, color = "blue")
```

```python
p=q = range(0,11)
d = range(0,3)
pdq = list(itertools.product(p,d,q))
```

```python
for param in pdq:
    try:
        model_arima=ARIMA(model_data, order=param)
        model_arima_fit = model_arima.fit()
        print(param, model_arima_fit.aic)
    except:
        continue
```

```python
model_arima1 = ARIMA(model_data, order=(1,2,3))
model_arima1_fit=model_arima1.fit()
predictions2 = model_arima1_fit.forecast(steps=10)[0]
print(predictions2)
plt.plot(predict_data, color="blue")
plt.plot(predictions2, color="red")
```
