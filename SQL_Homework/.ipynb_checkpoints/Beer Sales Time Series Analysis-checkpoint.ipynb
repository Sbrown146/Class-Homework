```python
from sqlalchemy import Table, MetaData, create_engine, select
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from scipy import stats
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima_model import ARIMA
```

```python
engine=create_engine("postgresql://postgres:postgres@localhost/BEERSALES")
with engine.connect() as conn:
    conn.execute("SET search_path TO 'BEER'")
    conn.execute("SET search_path TO 'BEER'").keys()
    meta=MetaData()
    Beer_table=Table("beer_sales", meta, autoload=True, autoload_with=conn)
    Beer_data=conn.execute(select([Beer_table])).fetchall()
    col=Beer_table.c
```

```python
print(Beer_table.c)
```

```python
Beer_df=pd.DataFrame(Beer_data, columns=[str(c) for c in Beer_table.c])
Beer_df=Beer_df.rename(columns={"beer_sales.beer_sales": "Beer Sales"})

Beer_df.head()
```

```python
# FFT in R would seem appropriate to try out here.

k1, p1 = stats.normaltest(Beer_df)
print(k1)
print(p1)

plt.figure(figsize=(20,5))
plt.plot((np.arange(0, Beer_df["Beer Sales"].count(),1)), Beer_df["Beer Sales"], marker="o")


plt.gca().set_yticklabels(['${:0,.0f}'.format(x) for x in plt.gca().get_yticks()])
plt.xlabel("Months")
plt.ylabel("Total Sales")
```

```python
plt.figure(figsize=(40,10))
print(plot_acf(Beer_df["Beer Sales"]))
print(plot_pacf(Beer_df))
```

```python
Beer_Diff=Beer_df.diff(periods=1)
Beer_Diff.head()
```

```python
plt.figure(figsize=(30,10))
plt.plot(np.arange(1,193,1), Beer_Diff, marker="o")
plt.show()
```

```python
k2, p2 = stats.normaltest(Beer_Diff)
print(k2)
print(p2)

plt.figure(figsize=(15,5))
plt.scatter(np.arange(1,193,1), Beer_Diff, marker="o")
```
